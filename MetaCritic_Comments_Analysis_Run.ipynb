```python
# imports
import ipywidgets as widgets
from IPython.display import display
from comment_analysis import comment_analysis
```

```python
# Aspects list
aspects_options = ['story', 'graphic', 'gameplay', 'difficulty', 'frames', 'server', 'multiplayer', 'boring']

# Widgets for input
game_name_widget = widgets.Text(
    description='Game Name:',
    placeholder='Type game name here',
    style={'description_width': 'initial'}
)

aspects_widget = widgets.SelectMultiple(
    options=aspects_options,
    description='Aspects:',
    value=[aspects_options[0]],  # Default value: story
    style={'description_width': 'initial'}
)

platform_widget = widgets.Text(
    description='Platform:',
    placeholder='Type platform name here',
    style={'description_width': 'initial'}
)

reviewer_widget = widgets.Dropdown(
    options=['user', 'critic'],
    description='Reviewer:',
    value='user',  # Default value
    style={'description_width': 'initial'}
)

# Integer slider with validation for multiples of 50
num_comments_widget = widgets.IntSlider(
    value=50,
    min=50,
    max=500,
    step=50,
    description='Num Comments:',
    continuous_update=False,
    style={'description_width': 'initial'}
)

# Create a button to trigger the function
run_button = widgets.Button(description="Run Analysis")

# Function to execute when button is clicked
def on_button_click(b):
    game_name = game_name_widget.value
    aspects = list(aspects_widget.value)
    platform = platform_widget.value
    reviewer = reviewer_widget.value
    num_comments = num_comments_widget.value

    # Call the comment_analysis function with the selected inputs
    comment_analysis(game_name, aspects, platform, reviewer, num_comments)

# Link the button click to the function
run_button.on_click(on_button_click)

# Display all widgets
display(game_name_widget, aspects_widget, platform_widget, reviewer_widget, num_comments_widget, run_button)
```
